# Q2

```python
def l2_dist(a, b):
    result = ((a - b) * (a - b)).sum()
    result = result ** 0.5
    return result
```

```python
import numpy as np
a = np.random.rand(20,20)
print(a)

b = np.random.rand(20,20)
print(b)
```

```python
l2_dist(a, b)
```

```python
l2_dist(a.T, b.T)
```

```python
l2_dist(np.reshape(a, (20 * 20)), np.reshape(b, (20 * 20)))
```

```python
l2_dist(np.reshape(a, (20 * 20)), np.reshape(b, (20 * 20, 1)))
```
